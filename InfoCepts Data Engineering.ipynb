## Data Processing

```python
sub = pd.read_csv(DATA_DIR+"infocept_submission.csv") 
```

```python
sub.info()
```

```python
# 0. Create 'is_special_day', 'prod_views_freqs' and 'prod_buys_freqs' columns.
#Better without thems !?
sub['is_special_day'] = data['is_special_day']
sub['prod_views_freqs'] = data['prod_views_freqs']
sub['prod_buys_freqs'] = data['prod_buys_freqs']
```

```python
# 1. Process the "start_time_ts" column to create "time_spec_points" column.
def time_points(x):  
    ans = pd.to_datetime(x).hour
    if ans >= 16:
        ans += 5
    
    return ans
 
sub["time_spec_points"] = data["start_time_ts"].apply(lambda x: time_points(x)) 
```

```python
# 2. Reduce the "total_duration"
sub["total_duration"] = data["total_duration"].apply(lambda x: round(x, 2))
```

```python
# 3.  Add the values of columns  "total_clicks", "total_items", "total_cats"
# to create a new column: "total_inventory".
sub["total_inventory"] = data.apply(lambda row: row.total_clicks + row.total_items + row.total_cats 
                                    if row.total_clicks + row.total_items + row.total_cats <= 10
                                    else (row.total_clicks + row.total_items + row.total_cats) * 2
                                    ,axis=1) 
```

```python
# 4. Create a new column "give_big_discount"
sub["give_big_discount"] = data.apply(lambda row: 1 if  row.day_of_week == 0 and row.is_special_day == 1 else 0, axis=1) 
```

```python
# 5. Create a new column prod_views_buys_ratio 
sub["prod_views_buys_ratio"] = data.apply(lambda row: row["prod_views_freqs"] / row["prod_buys_freqs"], axis=1) 
```

```python
# 6. Create a new column loyalty_points
def loyalty(row):  
    ans = 0
    if row.day_of_week == 0 and row.total_duration > 3: # Sunday?
        ans = 10*(row.total_duration - 3)
    return ans
 
sub["loyalty_points"] = data.apply(lambda row: loyalty(row), axis=1) 
```
